```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import random
```

```python
# Load the MNIST dataset
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data() # Splitting it into training and testing
import matplotlib.pyplot as plt

# Display the second training image using matshow
plt.matshow(x_train[1])
plt.show()

# Display the first training image using imshow
plt.imshow(-x_train[0], cmap="gray")
plt.show()

print(x_train.shape)
print(x_test.shape)
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
# c. DEFINE THE NETWORK ARCHITECTURE USING KERAS 
# Define the model
model = keras.Sequential([
    keras.Input(shape=(28, 28)),             # Define the input shape explicitly
    keras.layers.Flatten(),                  # Flatten the input images
    keras.layers.Dense(128, activation="relu"),  # Add a dense layer with 128 neurons and ReLU activation
    keras.layers.Dense(10, activation="softmax") # Add the output layer with 10 neurons and softmax activation
])


# Display the model summary
model.summary()
```

```python
# d. TRAIN THE MODEL USING SGD -&gt;
model.compile(optimizer="sgd",
loss="sparse_categorical_crossentropy",
metrics=['accuracy'])

history=model.fit(x_train,
y_train,validation_data=(x_test,y_test),epochs=10)
```

```python

# Evaluate the network
test_loss, test_acc = model.evaluate(x_test, y_test)

# Print the results
print("Loss = %.3f" % test_loss)
print("Accuracy = %.3f" % test_acc)
```
